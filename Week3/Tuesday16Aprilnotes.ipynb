create a very simple plot

```python
import numpy as np #imports. IPy notebook already does this though
t = arange(0.0,1.0+0.01,0.01) #+0.01 so 1.00 inclusive
c = cos(2*2*pi*t)
s = sin(2*2*pi*t)
```

```python
print pi
```

```python
#matplotlib is similar to matlab
#every plot must have axis lables
plot(t,s, color = 'red', linewidth = 17, linestyle = ':', label = "$cos(4{\pi}t)$") #\ is escape character. like \sigma or \Sigma
plot(t, c, color = 'blue', linewidth = 17, linestyle = '--', label = "$cos(4{\pi}t)$") #linestyle, linewidty, lable, color all modify the plot. sequence isnt important
xlabel('time (s)', fontsize = 10) #labels x axis
ylabel('voltage (mV)') #lables y axis
title('A simple plot', fontsize=20)
grid(True)
legend(loc="lower left")#default is upper right
savefig('simple_plot.pdf') #can save as pdf, jpeg, mnp
```
